for 600m and 1.3b

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M") #facebook/nllb-200-1.3B
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
```

```python
pip install datasets
```

```python
from datasets import load_dataset

dataset = load_dataset("LibrAI/do-not-answer")
```

```python
len(dataset['train']['question'])
dataset['train']['question']
```

```python
import random
random_numbers = random.sample(range(939), 100)


questions = [dataset['train']['question'][i] for i in random_numbers]


questions
```

```python
len(questions)
```

```python
inputs = tokenizer(questions, return_tensors="pt", padding = True)
```

```python
translated_tokens = model.generate(
    **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["zul_Latn"]
)
zulu = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)
```

```python
len(zulu)
```

```python
zulu
```

```python
translated_tokens = model.generate(
    **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["ben_Beng"]
)
bengali = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)
```

```python
translated_tokens = model.generate(
    **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["hat_Latn"]
)
haiti = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)
```

```python
import pandas as pd
df = pd.DataFrame(data = {'questions': questions, 'zulu': zulu, 'bengali': bengali, 'haiti': haiti})
df.to_csv('translated.csv', index=False)
```

```python
import requests

API_URL = "https://d43ht0l3h4sfbrwj.us-east-1.aws.endpoints.huggingface.cloud"  # Replace this
headers = {
    "Authorization": "Bearer hf_xxx",  # Replace this
    "Content-Type": "application/json"
}


def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()
```

```python
zulu = []
```

```python
df = pd.read_csv('/content/translated.csv')
questions = df['questions'].tolist()
```

```python
for question in questions:
  def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()


  output = query({"inputs": question, "parameters": {"src_lang": "eng_Latn", "tgt_lang": "zul_Latn"}})
  zulu.append(output[0]['translation_text'])
  print(output[0]['translation_text'])
```

```python
bengali = []
for question in questions:
  def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()


  output = query({"inputs": question, "parameters": {"src_lang": "eng_Latn", "tgt_lang": "ben_Beng"}})
  bengali.append(output[0]['translation_text'])
  print(output[0]['translation_text'])
```

```python
haiti = []
for question in questions:
  def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()


  output = query({"inputs": question, "parameters": {"src_lang": "eng_Latn", "tgt_lang": "hat_Latn"}})
  haiti.append(output[0]['translation_text'])
  print(output[0]['translation_text'])
```

```python
df = pd.DataFrame(data = {'questions': questions, 'zulu': zulu, 'bengali': bengali, 'haiti': haiti})
df.to_csv('translated.csv', index=False)
```
