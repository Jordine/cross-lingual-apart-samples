```python
import evaluate
import csv
import nltk

# Load the BLEU metric
bleu = evaluate.load("bleu")

# Open the CSV file
with open('/content/retranslated-600m-distilled.csv', 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    bleu_predictions = []
    bleu_references = []

    # Loop through each row in the CSV file
    for row in csv_reader:
        translation = row['questions']
        translated = row['haiti']

        translation_tokens = nltk.word_tokenize(translation)
        translated_tokens = nltk.word_tokenize(translated)

        bleu_predictions.append(' '.join(translation_tokens))
        bleu_references.append([' '.join(translated_tokens)])

    # Calculate the BLEU score for each prediction-reference pair
    total_bleu_score = 0
    for prediction, reference in zip(bleu_predictions, bleu_references):
        bleu_score = bleu.compute([{'predictions': prediction, 'references': reference}])
        print(bleu_score['bleu'])
        total_bleu_score += bleu_score['bleu']

    # Calculate the average BLEU score
    avg_bleu_score = total_bleu_score / len(bleu_predictions)

    # Print the average BLEU score
    print("Average BLEU score:", avg_bleu_score)
```
